```python
# Importei a biblioteca base64 para codificar e descodificar as senhas do usuario
import base64
# Essa biblioteca é para o Python se comunicar com o sistema operacional
import os

# Esse arquivo será o nosso banco de dados e vai armazenar todos os usuarios cadastrados
ARQUIVO_BD_USUARIOS = "usuarios_iesb.txt"

def codificar_a_senha(senha):
    """Essa função codifica a senha em base64 para que não seja salva em texto puro"""
    return base64.b64encode(senha.encode('utf-8')).decode('utf-8')

def descodificar_a_senha(senha_codificada):
    """Essa função descodifica a senha base64 para comparação no login"""
    return base64.b64decode(senha_codificada.encode('utf-8')).decode('utf-8')


#Aqui vai ficar as funções principais do sistema de autentificação de usuários

def cadastrar_usuario_iesb():
    """Essa função vai cadastrar um novo usuário e verificar se já existe"""
    print("\n"+"-"*35)
    print("IESB - Cadastre-se em Aluno Online")
    print("-"*35)

    email_usuario = input("Digite seu email: ")

    #Vai verificar se o usuario já existe no sistema
    usuario_existente = {} #Cria um dicionário vazio
    if os.path.exists(ARQUIVO_BD_USUARIOS):
        with open(ARQUIVO_BD_USUARIOS, "r") as arquivo:
            for linha in arquivo:
              # É para ignorar as linhas que estão em branco
                if not linha.strip():
                    continue
                u, _ = linha.strip().split(";")
                usuario_existente[u] = True

    #Se já existir mostrará a mensagem de erro
    if email_usuario in usuario_existente:
        print("Erro ao cadastrar!! Esse usuário já está cadastrado")
        return

    #Caso não exista pede para cria uma senha
    senha_usuario = input("Crie uma senha: ")
    #Aqui vai codificar a senha do usuario antes de salvar no arquivo dos usuários
    senha_codificada = codificar_a_senha(senha_usuario)

    #Vai agora adicionar ao arquivo o  usuario
    with open(ARQUIVO_BD_USUARIOS, "a") as arquivo:
        arquivo.write(f"{email_usuario};{senha_codificada}\n")

    #Confirma o cadastro
    print("\nUsuário cadastrado com sucesso!")
    print("-"*35)


def login_usuario_iesb():
    """Essa função vai autenticar o usuário e permitir login"""
    print("\n"+"-"*35)
    print("IESB - Login Aluno Online")
    print("-"*35)

    #Solicita o email e a senha do usuario
    email_usuario = input("Digite seu email: ")
    senha_usuario = input("Digite sua senha: ")

    #Vai verificar se arquivo de usuarios existe, se o aruivo não existe ninguem foi cadastrado ainda
    if not os.path.exists(ARQUIVO_BD_USUARIOS):
        print("Erro ao fazer login!! Usuário não cadastrado")
        return

  #Para controle se o login deu certo ou não
    sucesso_login = False


    #Aqui vai abrir o arquivo e lê os usuario por usuario para achar o cadastro do usario existente
    with open(ARQUIVO_BD_USUARIOS, "r") as arquivo:
        for linha in arquivo:
           # É para ignorar as linhas que estão em branco
            if not linha.strip():
                continue

            usuario_salvo, senha_salva = linha.strip().split(";")
            #Vai descodifiacar a senha que está codificada
            senha_salva = descodificar_a_senha(senha_salva)

            if email_usuario == usuario_salvo and senha_usuario == senha_salva:
                sucesso_login = True #Significa que o login foi validado
                break

 #Mensagem de login realizado com sucesso para mostrar para o usuario
    if sucesso_login:
        print("\nLogin realizado com sucesso! Bem-vindo(a)")
        print("-"*35)
    else:
        print("\nErro!! Usuário ou senha incorretos") #Erro caso não existir
        print("-"*35)


def opcaos_autenticacao_usuarios():
  """
  Essa função vai exibir o menu de opções para o usuário
  """
  print("\n"+"-"*35)
  print("IESB - Menu de Autenticação")
  print("-"*35)

  #Exibir opções ao usuario
  print("1 - Primeiro acesso? Cadastre-se")
  print("2 - Login")
  print("3 - Sair")


  #Vai solicitar a opção que ele deseja e vai armazenar a opção
  opcao_escolhida = input("Digite a opção desejada: ")


  if opcao_escolhida == "1":
    cadastrar_usuario_iesb()
  elif opcao_escolhida == "2":
    login_usuario_iesb()
  elif opcao_escolhida == "3":
    print("Saindo do programa!! Até logo!")
    return

  else:
    print("Opção inválida!! Tente novamente")
    opcaos_autenticacao_usuarios()

#Após escolher executar a opção que o usuario escolher, ele executa as opções novamente
  opcaos_autenticacao_usuarios()

#Aqui vai fazer o menu ser executado
if __name__ == "__main__":
  opcaos_autenticacao_usuarios()









```
